### if statement
- 음료수 자판기 만들기

- 1번 누르면 콜라
- 2번 누르면 사이다
- 3번 누르면 환타
- 4번 누르면 취소
- 그외 번호 누르면 잘못된 번호입니다.

```python
selected = int(input())

if selected == 1:
  print('콜라를 선택하셨습니다')
elif selected == 2:
  print('사이다를 선택하셨습니다')
elif selected == 3:
  print('환타를 선택하셨습니다')
elif selected == 4:
  print('취소를 선택하셨습니다')
else:
  print('잘못된 번호입니다.')
```

```python
## Assignment
## 
## 금액까지 추가해서
## 해당 콜라는 500, 사이다는 500, 환타는 300
## 내가 1,000원을 넣었을 때
## 전체 예산에서 구매 후에 남은 금액을 잔돈으로 표시해주고
## 예산을 넘어간 경우는 금액을 초과하였다고 alert 해주기
##
beverage_names = ['콜라', '사이다', '환타']
beverage_price = [500, 500, 300]


insert_coin = int(input('금액을 투입하십시오 : '))
print(f'투입된 금액은 {insert_coin}원 입니다.')
remain_coin = insert_coin


n_beverage = len(beverage_names)
selected = n_beverage

while selected:
  print('메뉴 선택 =>', end=' ')
  for i in range(n_beverage):
    print(f'{beverage_names[i]}({i+1}) : {beverage_price[i]}', end=' | ')
  print('반환(0)')  
  
  selected = int(input('메뉴를 선택해주십시오 : '))
  if selected == 0:
    break
  
  if remain_coin < beverage_price[selected-1]:
    print('잔액이 부족합니다.')
    continue
  else:
    remain_coin -= beverage_price[selected-1]
    print(f'선택된 음료는 {beverage_names[selected-1]} 입니다.')
    print(f'남은 잔액은 {remain_coin} 입니다.')

print(f'반환된 금액은 {remain_coin} 입니다.')
    
```

```python

beverage_names = ['콜라', '사이다', '환타']
beverage_price = [500, 500, 300]


insert_coin = int(input('금액을 투입하십시오 : '))
print(f'투입된 금액은 {insert_coin}원 입니다.')
remain_coin = insert_coin


n_beverage = len(beverage_names)
selected = n_beverage

while selected:
  if not any(remain_coin >= price for price in beverage_price):
    break
  
  print('메뉴 선택 =>', end=' ')
  for i in range(n_beverage):
    if (remain_coin >= beverage_price[i]):
      print(f'{beverage_names[i]}({i+1}) : {beverage_price[i]}', end=' | ')
  print('반환(0)')  
  
  selected = int(input('메뉴를 선택해주십시오 : '))
  if selected == 0:
    break
  
  if remain_coin < beverage_price[selected-1]:
    print('잔액이 부족합니다.')
    continue
  else:
    remain_coin -= beverage_price[selected-1]
    print(f'선택된 음료는 {beverage_names[selected-1]} 입니다.')
    print(f'남은 잔액은 {remain_coin} 입니다.')

print(f'반환된 금액은 {remain_coin} 입니다.')
    
```

#### 중첩루프

- for i in range()

```python
for i in range(8):
  for j in range(2):
    print(j, sep=' ', end=' ')
  print(i, sep =' ')
```

```python
for i in range(2):
  for j in range(8):
    print('안쪽', j, sep=' ', end = '')
    print('바깥', i, end=' ')
```

```python
for i in range(2):
  for j in range(3):
    for k in range(4):
      print(k, end=' ')
    print(j, end=' ')
  print(i, end=' ')
```
